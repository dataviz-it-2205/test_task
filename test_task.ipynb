```python
import pandas as pd

xls = pd.read_excel("tasks.xlsx", sheet_name=['Данные об аудитории', 'Данные АБ тестов', 'Листеры'])

df1 = xls['Данные об аудитории']
df2 = xls['Данные АБ тестов']
df3 = xls['Листеры']
```

```python
# 1

df1['date'] = pd.to_datetime(df1['date'])

november_data = df1[df1['date'].dt.month == 11]

mau = november_data['user_id'].nunique()

print("MAU за ноябрь:", mau)
```

```python
# 2

dau_per_day = df1.groupby('date')['user_id'].nunique()

print("DAU по дням:")
print(dau_per_day.mean())
```

```python
# 3
new_users_nov1 = df1[df1['date'] == '2023-11-01']['user_id'].unique()
returning_users_nov2 = df1[(df1['date'] == '2023-11-02') & (df1['user_id'].isin(new_users_nov1))]['user_id'].unique()
retention = (len(returning_users_nov2) / len(new_users_nov1)) * 100
print(f"First Day Retention: {retention:.2f}%")
```

```python
# 5
total_users_november = df1['user_id'].nunique()
users_with_view_adverts = df1[df1['view_adverts'] > 0]['user_id'].nunique()
conversion = (users_with_view_adverts / total_users_november) * 100
print(f"Conversion: {conversion:.2f}%")
```

```python
# 6
total_views = df1['view_adverts'].sum()
total_users_november = df1['user_id'].nunique()
average_views_per_user = total_views / total_users_november
print(f"Average number of ads viewed per user in November: {average_views_per_user:.1f}")
```

```python
# 7
total_users = 2000
detractors = 500
promoters = 1200

percent_promoters = (promoters / total_users) * 100
percent_detractors = (detractors / total_users) * 100

nps = percent_promoters - percent_detractors
print(f"NPS: {nps}%")
```

```python
# 8
from scipy import stats
control_group = df2[df2['experiment_group'] == 'control']['revenue']
test_group = df2[df2['experiment_group'] == 'test']['revenue']

t_stat, p_value = stats.ttest_ind(test_group, control_group)

print(f"p-value for experiment 1: {p_value}")
```

```python
# 9
average_revenue_per_user = df3['revenue'].sum() / df3['user_id'].nunique()
print(f"Average revenue per user: {average_revenue_per_user:.2f}")
```

```python
# 10
median_age = df3['age'].median()
print(f"Median age: {median_age}")
```
